````python
from sklearn.datasets import load_iris
iris = load_iris()

x_names = iris.feature_names
y_names = iris.target_names

data1 = iris.data
print('x data =', data1.shape)
y = iris.target
print('y data =', y.shape)


df = pd.DataFrame(data1)
df.columns = x_names
df.head()
````

````python
df['species'] = y
df.head()
````

#### | 데이터 기본 탐색


````python
sns.relplot(data=df)
````

### (2) Distributions plot


````python
sns.displot(data = df, x = 'bill_length_mm', y = 'bill_depth_mm',kind = 'hist', hue = 'species')
````

### (3) Joint plot


````python
joint = sns.jointplot(x = 'body_mass_g', y = 'flipper_length_mm', data = df,
                     color = 'green' )
````

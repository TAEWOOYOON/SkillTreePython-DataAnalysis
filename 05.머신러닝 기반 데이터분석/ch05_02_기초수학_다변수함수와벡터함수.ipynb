#### **ÏòàÏ†ú 05**

\begin{align}

#### **ÏòàÏ†ú 07**

\begin{align}
f(t) = \begin{pmatrix}
x(t)\\ 
y(t)\\ 
z(t)\\ \end{pmatrix} = \begin{pmatrix}
sin(6t)\\
\frac{t}{4}\\
\frac{t^2}{2}\\
\end{pmatrix}

* Ìï®Ïàò ÏÉùÏÑ±ÌïòÍ∏∞

````python
def f7(t) :
  return np.array([np.sin(6*t),t/4,(t**2)/2])
````

* Ïù∏Ìíã Ï†ïÏùò (-1~3 200Í∞ú)

````python
xv = np.linspace(-1,3,200)
A,B,C = f7(xv)
xv.shape, A.shape, B.shape, C.shape
````

````python
fig = plt.figure(figsize = (10,5))
ax = plt.axes(projection = '3d')
ax.scatter3D(A,B,C, color = 'g')

ax.set_title('f(x),f(y),f(z)')
ax.set_xlabel('sin(6t)')
ax.set_ylabel('$t/4$')
ax.set_zlabel('$(t**2)/2$')
````

* Í∑∏ÎûòÌîÑ ÏÑ§Î™Ö
  * tÍ∞Ä Í∞ÄÏßàÏàò ÏûàÎäî Í∞í (Ïù∏Ìíã) : Î™®Îì† Ïã§Ïàò
  * x(t)Í∞Ä Í∞ÄÏßà Ïàò ÏûàÎäî Í∞í (Ï≤´Î≤àÏß∏ ÏïÑÏõÉÌíã) : -1 ~ 1 ÏÇ¨Ïù¥Ïùò Í∞í
  * y(t)Í∞Ä Í∞ÄÏßà Ïàò ÏûàÎäî Í∞í (ÎëêÎ≤àÏß∏ ÏïÑÏõÉÌíã) : Î™®Îì† Ïã§Ïàò
  * z(t)Í∞Ä Í∞ÄÏßà Ïàò ÏûàÎäî Í∞í (ÏÑ∏Î≤àÏß∏ ÏïÑÏõÉÌíã) : 0 Ïù¥ÏÉÅÏùò Ïã§Ïàò

#### **ÏòàÏ†ú 08**

\begin{align}
\tau(t) = \begin{pmatrix}
x(t)\\ 
y(t)\\ 
z(t)\\ \end{pmatrix} = \begin{pmatrix}
cos(t)\\
sin(t)\\
t\\
\end{pmatrix}
\end{align}

* Ìï®Ïàò ÏÉùÏÑ±ÌïòÍ∏∞

````python
def tau(t) :
  return np.array([np.cos(t), np.sin(t), t])
````

* Ïù∏Ìíã Ï†ïÏùò(-10 ~ 10 ÏÇ¨Ïù¥ 200Í∞ú)

````python
xv = np.linspace(-10,10,200)
x,y,z = tau(xv)
print(xv.shape, x.shape, y.shape, z.shape)
````

* 3dÌîåÎ°ØÏúºÎ°ú ÏãúÍ∞ÅÌôî

````python
fig = plt.figure(figsize = (10,5))
ax = plt.axes(projection = '3d')
ax.scatter3D(x,y,z, color = 'g')

ax.set_title('tau(t)')
ax.set_xlabel('cos(t)')
ax.set_ylabel('sint(t)')
ax.set_zlabel('t')
````

* Í∑∏ÎûòÌîÑ ÏÑ§Î™Ö

  * cos(t)Í∞Ä Í∞ÄÏßà Ïàò ÏûàÎäî Í∞í (Ï≤´Î≤àÏß∏ ÏïÑÏõÉÌíã) : -1 ~ 1 ÏÇ¨Ïù¥Ïùò Î™®Îì†Í∞í
  * sin(t)Í∞Ä Í∞ÄÏßà Ïàò ÏûàÎäî Í∞í (ÎëêÎ≤àÏß∏ ÏïÑÏõÉÌíã) : -1 ~ 1 ÏÇ¨Ïù¥Ïùò Î™®Îì†Í∞í

#### **ÏòàÏ†ú 09**


\begin{align}

y^2\\

\end{align}

* Ìï®Ïàò ÏÉùÏÑ±ÌïòÍ∏∞

````python
def f9(x,y) :
  return np.array([x**2, y**2, x*y])

f9(1,2)
````

````python
xv = np.linspace(-0.5,1,50)
yv = np.linspace(-0.5,1,50)

# x,y Í≤©Ïûê ÏàúÏÑúÏåç
X,Y = np.meshgrid(xv,yv)
# ÏïÑÏõÉÌíã Ï†ïÏùò
A,B,C = f9(X,Y)
X.shape, Y.shape, A.shape, B.shape, C.shape
````

````python
fig = plt.figure(figsize = (10,5))
ax = plt.axes(projection = '3d')  # 3Ï∞®Ïõê ÌîåÎ°Ø
ax.scatter3D(A,B,C, marker = '.',color = 'g')
ax.plot_surface(A,B,C, alpha = .4)
ax.set_title('f(x,y) according to x,y')
ax.set_xlabel('$x^2$')
ax.set_ylabel('$y^2$')
ax.set_zlabel('$xy$')
plt.show()
````

u\\ 
v\\ 
1+u^2+\frac{v}{1+v^2}\\ \end{pmatrix}, \ \ \ \ \ 0 \leq u, v \leq1
\end{align}

* Ìï®Ïàò ÏÉùÏÑ±ÌïòÍ∏∞

````python
def f10(u,v) :
  return np.array([u,v,1+u**2+v/(1+v**2)])
````

* Ïù∏Ìíã Ï†ïÏùò (u, v Î™®Îëê 0 ~ 1 ÏÇ¨Ïù¥ 50Í∞ú)

````python
uv = np.linspace(0,1,50)
vv = np.linspace(0,1,50)

U, V = np.meshgrid(uv,vv)

x,y,z = f10(U,V)
print(U.shape, V.shape, x.shape, y.shape, z.shape)
````

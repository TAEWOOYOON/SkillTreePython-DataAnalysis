### 일변수 스칼라 함수
---

하나의 인풋(=스칼라, 일변수), 하나의 아웃풋(=스칼라, 스칼라함수)

\begin{align}
y = f(x)
\end{align}

#### **예제 01**

\begin{align}
f(x) = x^2+2x+1
\end{align}

* 함수 생성하기

````python
def f1(x) : # x : 스칼라값
  return x**2+2*x+1

f1(-1)
````

* 인풋 정의 (-10~10, 100개)

````python
# 100개의 벡터
xv = np.linspace(-10,10,100)

# 100개의 스칼라값(xx의 성분)에 대한 100개 스칼라 아웃풋
fv = f1(xv)
xv.shape, fv.shape
````

* 2d플롯으로 시각화

````python
fig, ax = plt.subplots(1)
ax.plot(xv, fv)

# 설명
ax.set_title('f(x) according to x')  # 제목 설정
ax.set_xlabel('x')  # x축 라벨
ax.set_ylabel('f(x)')  # y축 라벨

x = 5.55
ax.plot(x,f1(x), marker = 'o', c= 'r')  # 특정 x에 따른 f(x) 표시
ax.text(x, f1(x), f'({x},{np.round(f1(x))})')  # np.round : 소수점 아래 두자리까지 표시
````

* 그래프 설명
  * x가 가질수 있는 값 : 모든 실수
  * f(x)가 가질 수 있는 값: 0 이상의 모든 실수
  * 그래프의 특징 : x가 -1일 때 아웃풋이 가장 작고 x가 -1에서 멀어질수록 아웃풋이 점점 커진다.

### 다변수 스칼라 함수
---

여러개의 인풋(=다변수), 하나의 아웃풋(=스칼라, 스칼라함수)


\begin{align}
y = f(𝐱)\\
\end{align}



#### **예제 03**

\begin{align}
f(x, y) = x+y
\end{align}

* 함수 생성하기

````python
def f3(x, y) :
  return x+y

f3(2,2)
````

* 인풋 정의 (x, y, 모두 -15~20 값)

````python
xv = np.linspace(-15,25,40)
yv = np.linspace(-15,25,40)
print(xv.shape)
print(yv.shape)
X,Y = np.meshgrid(xv,yv)    # np.meshgrid(a,b)는 a와 b의 배열을 각각 하나의 성분씩 매치해주는 함수. (a =1 b=1,2,3,4,5,....., a=2 b=1,2,3......) -> 값이 두개의 행렬이 나옴.
f3(X,Y).shape
````

* 3d플롯으로 시각화

````python
fig = plt.figure(figsize = (9,7))
ax = plt.axes(projection = '3d') # 3차원 플롯
ax.scatter3D(X,Y, f3(X,Y)) # x,y,z 축 데이터를 입력 -> X, Y값에 따른 함수값을 보겠다는 뜻

# 설명
ax.set_title('f(x,y) according to x,y')  # 제목 설정
ax.set_xlabel('x')  # x축 라벨
ax.set_xlabel('y')  # y축 라벨
ax.set_ylabel('z')  # z축 라벨

x = 5.55
y = 10
ax.scatter3D(x,y,f3(x,y), marker = 'o', c= 'r')  # 특정 x에 따른 f(x) 표시
ax.text(x, y,f3(x,y), f'({x},{y},{np.round(f3(x,y))})')  # np.round : 소수점 아래 두자리까지 표시
````

````python
fig = plt.figure(figsize = (9,7))
ax = plt.axes(projection = '3d') # 3차원 플롯
ax.scatter3D(X,Y, f3(X,Y)) # x,y,z 축 데이터를 입력 -> X, Y값에 따른 함수값을 보겠다는 뜻
surf = ax.plot_surface(X,Y,f3(X,Y), cmap = 'rainbow', alpha = .7) # 점을 이은 면 surface 그리기

# 디테일
ax.set_title('f(x,y) according to x,y')  # 제목 설정
ax.set_xlabel('x')  # x축 라벨
ax.set_xlabel('y')  # y축 라벨
ax.set_ylabel('z')  # z축 라벨

# 컬러바
fig.colorbar(surf, label = 'f(x,y)', shrink = 0.6)
plt.show()
````

* 그래프 설명
  * x, y가 가질수 있는 값 (인풋) : 모든 실수
  * f(x)가 가질 수 있는 값 (아웃풋) : 모든 실수
  * 그래프의 특징 : 최소값이 없다. x,y 가 작아질수록 f(x,y)의 값도 계속 작아진다.

#### **예제 04**

\begin{align}
f(x, y) = (x-5)^2+(y-5)^2 
\end{align}

* 함수 생성하기

````python
def f4(x,y) :
  return (x-5)**2 + (y-5)**2
````

````python
xv = np.linspace(-15,20,60)
yv = np.linspace(-15,20,60)

X,Y = np.meshgrid(xv,yv)
f4(X,Y).shape
````

* 그래프 설명
  * x, y가 가질수 있는 값 (인풋) : 모든 실수
  * f(x)가 가질 수 있는 값 (아웃풋) : 0이상의 모든 실수

## 일변수 벡터함수
---

하나의 인풋(=스칼라, 일변수), 여러개의 아웃풋(=스칼라, 스칼라함수)




\begin{align}
f(x) = (f_1(x), f_2(x), ..., f_n(x))^T
\end{align}

#### **예제 06**

\begin{align}
f(x) = \begin{pmatrix}
h(x)\\ 
g(x)\\ 
\end{pmatrix}  = \begin{pmatrix}
x^2\\ 
5x\\ 
\end{pmatrix}
\end{align}

* 함수 $f(x)$ 생성하기

````python
def f6(x) :
  return np.array([x**2,5*x])
````

* 인풋 정의 (-10~10 20개)

````python
xv = np.linspace(-10,10,20) # x값 20개
hv,gv = f6(xv)
# 인풋     첫번째 벡터(아웃풋)요소    두번째 벡터 요소
xv.shape,   hv.shape,                   gv.shape
````

* 2d플롯으로 시각화

````python
fig, ax = plt.subplots(1)
ax.plot(hv,gv) # x축을 h(x), y축을 g(x)

# 설명
ax.set_title('f(x) space')  # 제목 설정
ax.set_xlabel('h(x)')  # x축 라벨
ax.set_ylabel('g(x)')  # y축 라벨

scat = ax.scatter(hv,gv, c= xv, cmap = 'rainbow') # x축을 h(x), y축을 g(x)로 설정
fig.colorbar(scat)   # x값에 따라 컬러바 설정
plt.show
````

* 그래프 설명
  * x가 가질수 있는 값 (인풋) : 모든 실수
  * g(x)가 가질 수 있는 값 (첫번째 아웃풋) : 모든 실수 
  * h(x)가 가질수 있는 값 (두번째 아웃풋): g(x)가 0일때 최소값이고 모든 값이 0보다 크다

#### **예제 07**

\begin{align}
f(t) = \begin{pmatrix}
x(t)\\ 
y(t)\\ 
z(t)\\ \end{pmatrix} = \begin{pmatrix}
sin(6t)\\
\frac{t}{4}\\
\frac{t^2}{2}\\
\end{pmatrix}
\end{align}

* 함수 생성하기

````python
def f7(t) :
  return np.array([np.sin(6*t),t/4,(t**2)/2])
````

* 인풋 정의 (-1~3 200개)

````python
xv = np.linspace(-1,3,200)
A,B,C = f7(xv)
xv.shape, A.shape, B.shape, C.shape
````

* 3d플롯으로 시각화

````python
fig = plt.figure(figsize = (10,5))
ax = plt.axes(projection = '3d')
ax.scatter3D(A,B,C, color = 'g')

ax.set_title('f(x),f(y),f(z)')
ax.set_xlabel('sin(6t)')
ax.set_ylabel('$t/4$')
ax.set_zlabel('$(t**2)/2$')
````

* 그래프 설명
  * t가 가질수 있는 값 (인풋) : 모든 실수
  * x(t)가 가질 수 있는 값 (첫번째 아웃풋) : -1 ~ 1 사이의 값
  * y(t)가 가질 수 있는 값 (두번째 아웃풋) : 모든 실수
  * z(t)가 가질 수 있는 값 (세번째 아웃풋) : 0 이상의 실수

#### **예제 08**

\begin{align}
\tau(t) = \begin{pmatrix}
x(t)\\ 
y(t)\\ 
z(t)\\ \end{pmatrix} = \begin{pmatrix}
cos(t)\\
sin(t)\\
t\\
\end{pmatrix}
\end{align}

* 함수 생성하기

````python
def tau(t) :
  return np.array([np.cos(t), np.sin(t), t])
````

* 인풋 정의(-10 ~ 10 사이 200개)

````python
xv = np.linspace(-10,10,200)
x,y,z = tau(xv)
print(xv.shape, x.shape, y.shape, z.shape)
````

* 3d플롯으로 시각화

````python
fig = plt.figure(figsize = (10,5))
ax = plt.axes(projection = '3d')
ax.scatter3D(x,y,z, color = 'g')

ax.set_title('tau(t)')
ax.set_xlabel('cos(t)')
ax.set_ylabel('sint(t)')
ax.set_zlabel('t')
````

* 그래프 설명
  * t가 가질수 있는 값 (인풋) : 모든 실수
  * cos(t)가 가질 수 있는 값 (첫번째 아웃풋) : -1 ~ 1 사이의 모든값
  * sin(t)가 가질 수 있는 값 (두번째 아웃풋) : -1 ~ 1 사이의 모든값


## 다변수 벡터함수
---
여러개의 인풋(=스칼라, 일변수), 여러개의 아웃풋(=스칼라, 스칼라함수)

\begin{align}
f(𝐱) = (f_1(𝐱), f_2(𝐱), ..., f_n(𝐱))^T
\end{align}

#### **예제 09**


\begin{align}
f(x,y) = \begin{pmatrix}
x^2\\ 
y^2\\ 
xy\\ \end{pmatrix}
\end{align}

* 함수 생성하기

````python
def f9(x,y) :
  return np.array([x**2, y**2, x*y])

f9(1,2)
````

* 인풋 정의 (x,y 모두 -0.5 ~ 1 사이 50개)

````python
xv = np.linspace(-0.5,1,50)
yv = np.linspace(-0.5,1,50)

# x,y 격자 순서쌍
X,Y = np.meshgrid(xv,yv)
# 아웃풋 정의
A,B,C = f9(X,Y)
X.shape, Y.shape, A.shape, B.shape, C.shape
````

* 3d플롯으로 시각화

````python
fig = plt.figure(figsize = (10,5))
ax = plt.axes(projection = '3d')  # 3차원 플롯
ax.scatter3D(A,B,C, marker = '.',color = 'g')
ax.plot_surface(A,B,C, alpha = .4)
ax.set_title('f(x,y) according to x,y')
ax.set_xlabel('$x^2$')
ax.set_ylabel('$y^2$')
ax.set_zlabel('$xy$')
plt.show()
````

#### **예제 10**


\begin{align}
f(u,v) = \begin{pmatrix}
u\\ 
v\\ 
1+u^2+\frac{v}{1+v^2}\\ \end{pmatrix}, \ \ \ \ \ 0 \leq u, v \leq1
\end{align}

* 함수 생성하기

````python
def f10(u,v) :
  return np.array([u,v,1+u**2+v/(1+v**2)])
````

* 인풋 정의 (u, v 모두 0 ~ 1 사이 50개)

````python
uv = np.linspace(0,1,50)
vv = np.linspace(0,1,50)

U, V = np.meshgrid(uv,vv)

x,y,z = f10(U,V)
print(U.shape, V.shape, x.shape, y.shape, z.shape)
````

* 3d플롯으로 시각화

````python
fig = plt.figure(figsize = (10,5))
ax = plt.axes(projection = '3d')  # 3차원 플롯
ax.scatter3D(x,y,z, marker = '.',color = 'g')
ax.set_title('f(u,v)')
ax.set_xlabel('u')
ax.set_ylabel('v')
ax.set_zlabel('$1+u^2+v/(1+v^2)$')
plt.show()
````

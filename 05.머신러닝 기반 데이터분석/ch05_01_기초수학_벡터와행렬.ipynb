### **| 벡터와 벡터의 연산**

\begin{split}

## 행렬
---

* Second-order Tensor Operation

\begin{split}
\begin{align}
A=
\begin{bmatrix}
a_{11} \ \ \ a_{12} \ \ \ ... \ \ \ a_{1M} \\
a_{21} \ \ \ a_{22} \ \ \ ... \ \ \ a_{2M} \\
... \ \ \ ... \ \ \ ... \ \ \ ... \\
... \ \ \ ... \ \ \ a_{ij} \ \ \ ... \\
... \ \ \ ... \ \ \ ... \ \ \ ... \\
a_{N1} \ \ \ a_{N2} \ \ \ ... \ \ \ a_{NM} \\
\end{bmatrix}
\ \ , \ \ A∈ℝ^{N\times M}
\end{align}
\end{split}




### **| 행렬과 행렬의 연산**


* 덧셈, 뺄셈

\begin{align}
A+(-)B=
\begin{bmatrix}
a_{11} \ \ \ a_{12} \ \ \ ... \ \ \ a_{1M} \\
a_{21} \ \ \ a_{22} \ \ \ ... \ \ \ a_{2M} \\
... \ \ \ ... \ \ \ ... \ \ \ ... \\
... \ \ \ ... \ \ \ a_{ij} \ \ \ ... \\
... \ \ \ ... \ \ \ ... \ \ \ ... \\
a_{N1} \ \ \ a_{N2} \ \ \ ... \ \ \ a_{NM} \\
\end{bmatrix}
+(-)
\begin{bmatrix}
b_{11} \ \ \ b_{12} \ \ \ ... \ \ \ b_{1M} \\
b_{21} \ \ \ b_{22} \ \ \ ... \ \ \ b_{2M} \\
... \ \ \ ... \ \ \ ... \ \ \ ... \\
... \ \ \ ... \ \ \ b_{ij} \ \ \ ... \\
... \ \ \ ... \ \ \ ... \ \ \ ... \\
b_{N1} \ \ \ b_{N2} \ \ \ ... \ \ \ b_{NM} \\
\end{bmatrix}
\end{align}
\begin{align}
\\
\\
A∈ℝ^{N\times M} \ \ \ \ B∈ℝ^{N\times M}

#### **예제 07**

\begin{split}
\begin{align}
A =
\begin{bmatrix}
1 \ \ 2 \ \ 3 \\
4 \ \ 5 \ \ 6 \\
\end{bmatrix}
B =
\begin{bmatrix}
1 \ \ 0 \ \ 0 \\
0 \ \ 1 \ \ 0 \\
0 \ \ 0 \ \ 1 \\
\end{bmatrix}
,\;\;
𝐱=
\begin{bmatrix}
0 \\
1 \\
2 \\
\end{bmatrix}
, \; c = 2
\end{align}
\end{split}




* $A, B, 𝐱, c$ 정의하기

````python
A = np.array( [ [1,2,3], 
               [4,5,6] ] )

B = np.array( [ [1,0,0], 
               [0,1,0],
               [0,0,1]] )
x = np.array([ [0],
              [1],
              [2]])
c = 2

print(A.shape, B.shape, x.shape)
````

* $cAB$ 을 계산하고 ℝ 의 변화를 쓰세요.

````python
a = c*np.dot(A,B)
print(a, a.shape)
# c는 스칼라값, A는 2X3, B는 3X3 이므로 c값은 R에 영향을 주지 않고 2X3 * 3X3 = 2X3 이 된다
````

* $AB𝐱$

````python
a = np.dot(np.dot(A,B),x)
print(a, a.shape)
# 2X3 * 3X3 * 3X1 -> 2X1
````

* $𝐱^TBA^T$ 을 계산하고 ℝ 의 변화를 쓰세요.

````python
a = np.dot(x.T, B)
b = np.dot(a,A.T)
print(b, b.shape)

# x.T : 1X3, B : 3X3, A.T : 3X2  이므로 1X3 * 3X3 * 3X2 = 1X2
````

#### **예제 08**

행렬 $A$가 다음과 같이 주어졌을 때 $AB$ 행렬이 각 shape를 만족하도록 $B$를 정의하세요. 단 $B$의 값은 1부터 5까지 랜덤 정수 입니다.

\begin{split}
\begin{align}
A =
\begin{bmatrix}
1 \ \ 1 \ \ 1 \ \ 1 \\
2 \ \ 2 \ \ 2 \ \ 2 \\
\end{bmatrix}
\end{align}
\end{split}






* $A$ 정의하기

````python
A = np.array( [ [1,1,1,1],
               [2,2,2,2]] )
A, A.shape
````

* $AB \in ℝ^{2\times5}$ 일 때 $B$를 정의하고 $AB$를 계산하세요.

````python
# R이 2X5 이므로 B의 열은 5 이다. A의 열과 B의 행이 같아야 하므로 B의 행은 4가 된다. 즉, B는 4X5 행렬이다.
B = np.random.randint(1,6,(4,5))

a = np.dot(A,B)
print(a,a.shape)
````

* $AB \in ℝ^{2\times3}$ 일 때 $B$를 정의하고 $AB$를 계산하세요.

````python
# R이 2X3 이므로 B의 열은 3 이다. A의 열과 B의 행이 같아야 하므로 B의 행은 4가 된다. 즉, B는 4X3 행렬이다.

B = np.random.randint(1,6,(4,3))

a = np.dot(A,B)
print(a,a.shape)
````

* $AB \in ℝ^{2\times1}$ 일 때 $B$를 정의하고 $AB$를 계산하세요.

````python
# R이 2X1 이므로 B의 열은 1 이다. A의 열과 B의 행이 같아야 하므로 B의 행은 4가 된다. 즉, B는 4X1 행렬이다.

B = np.random.randint(1,6,(4,1))

a = np.dot(A,B)
print(a,a.shape)
````

* $A$와 $M$ 두개의 값을 인풋으로 하면 $AB \in ℝ^{N\times M}$를 만족하는 랜덤행렬 $B$를 반환하도록 함수 `random_weights`를 생성하세요.

````python
def random_weights(A,M) :
  B = np.random.randint(0,10,(A.shape[1],M))
  return B
````

````python
 A = np.array( [ [1,2,3,4],
                [1,2,3,4],
                [5,6,7,8]] )
 random_weights(A, 5)
````

* 함수 사용 예시

````python
B = random_weights(A, 5)
print(A.shape, B.shape, (A@B).shape)
````

````python
 A = np.array( [ [1,2,3,4],
                [1,2,3,4],
                [5,6,7,8]] )
 B = random_weights(A, 5)
 print(A.shape, B.shape, np.dot(A,B))
````

* $A$ì™€ $M$ ë‘ê°œì˜ ê°’ì„ ì¸í’‹ìœ¼ë¡œ í•˜ë©´ $AB \in â„^{N\times M}$ë¥¼ ë§Œì¡±í•˜ëŠ” ëœë¤í–‰ë ¬ $B$ë¥¼ ë°˜í™˜í•˜ë„ë¡ í•¨ìˆ˜ `random_weights`ë¥¼ ìƒì„±í•˜ì„¸ìš”.

````python
def random_weights(A,M) :
  B = np.random.randint(0,10,(A.shape[1],M))
  return B
````
